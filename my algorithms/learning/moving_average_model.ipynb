# Moving Average Models
<p>The <strong>1st order moving average</strong> model</p>
$$x_t = \mu + w_t + \theta_{1}w_{t-1}$$

```python
from functools import partial
import math
import numpy as np
import scipy
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.stattools import jarque_bera
import pandas as pd

import matplotlib.pyplot as plt
```

<h3>1st Order Moving Average</h3>

```python
# simulate 1st order moving average model
def simulate_1st_order_ma_model(n, mu, theta):
    X = np.ndarray(n)
    W = np.ndarray(n)
    X[0] = mu
    W[0] = np.random.normal(0,1)
    
    # simulate an array of x's
    for i in range(1, n):
        W[i] = np.random.normal(0,1)
        X[i] = mu + W[i] + theta * W[i-1]
        
    return X, W
```

```python
X, _ = simulate_1st_order_ma_model(1000, 10, 0.7)
```

```python
# simulation plot
plt.plot(X)
plt.show()
```

```python
# autocorrelation plot
from pandas.plotting import autocorrelation_plot

data = pd.Series(X)
autocorrelation_plot(data)
plt.show()

print data.autocorr(1)
print data.autocorr(2)
print data.autocorr(3)
print data.autocorr(4)
print data.autocorr(5)
```
